```python
# Set dependencies
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plot
import scipy.stats as st
from sqlalchemy import create_engine
from sqlalchemy.orm import Session
```

```python
# Import csv file
data_path = "Data/WorldCupMatches.csv"
```

```python
# Read WorldCupMatches.csv into Pandas
soccer1_df = pd.read_csv(data_path)
soccer1_df.head()
```

```python
# Drop Datetime column
soccer1_df = soccer1_df.drop(soccer1_df.columns[[1]], axis=1) 
soccer1_df
```

```python
# Import csv
data_path = "Data/WorldCupPlayers.csv"
```

```python
# Read WorldCupPlayers.csv into Pandas
soccer2_df = pd.read_csv(data_path)
soccer2_df.head()
```

```python
# Create dataframe that shows the Year, Stage, Home Team, Away Team, Attendance and MatchID
visualization1_df = soccer1_df[["Year", "Stage", "Home Team Name", "Away Team Name", "Attendance", "MatchID"]].dropna()
visualization1_df
```

```python
# Rename columns
renamed_visualization1_df = visualization1_df.rename(columns={'Home Team Name': 'Home_Team_Name', 'Away Team Name': 'Away_Team_Name'})
renamed_visualization1_df
```

```python
# Group data by MatchID
grouped_visualization1_df = renamed_visualization1_df.groupby(["MatchID"]).sum()
grouped_visualization1_df
```

```python
# Create dataframe that shows the Year, Home Team Goals, Away Team Goals and City
visualization2_df = soccer1_df[["Year", "Home Team Goals", "Away Team Goals", "City"]].dropna()
visualization2_df
```

```python
# Rename columns
renamed_visualization2_df = visualization2_df.rename(columns={'Home Team Goals': 'Home_Team_Goals', 'Away Team Goals': 'Away_Team_Goals'})
renamed_visualization2_df
```

```python
# Group data by year and city
grouped_visualization2_df = renamed_visualization2_df.groupby(["Year", "City"]).sum()
grouped_visualization2_df
```

```python
# Import csv
data_path = "Data/WorldCups.csv"
```

```python
# Read WorldCups.csv into Pandas
soccer3_df = pd.read_csv(data_path)
soccer3_df.head()
```

```python
# Create dataframe that shows the Winning Country, Year and Goals Scored
visualization3_df = soccer3_df[["Winner", "Year", "GoalsScored"]].dropna()
visualization3_df
```

```python
# Connect to local database
engine = create_engine("postgresql://postgres:password@localhost:5433/Soccer")
connection = engine.connect()
```

```python
# Confirm table names in SQL
engine.table_names()
```

```python
# Add the Visualization 1 Dataframe
renamed_visualization1_df.to_sql("Vizualization1", connection, if_exists="replace")
```

```python
# Add the Visualization 2 Dataframe
grouped_visualization2_df.to_sql("Vizualization2", connection, if_exists="replace")
```

```python
# Add the Visualization 3 Dataframe
visualization3_df.to_sql("Vizualization3", connection, if_exists="replace")
```
