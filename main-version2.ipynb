```python
# Import csv
data_path = "Data/WorldCupPlayers.csv"
```

```python
# Read WorldCupPlayers.csv into Pandas
soccer2_df = pd.read_csv(data_path)
soccer2_df.head()
```

```python
# Create dataframe that shows the Year, Stage, Home Team, Away Team, Attendance and MatchID
visualization1_df = soccer1_df[["Year", "Stage", "Home Team Name", "Away Team Name", "Attendance", "MatchID"]].dropna()
visualization1_df
```

```python
# Rename columns
renamed_visualization1_df = visualization1_df.rename(columns={'Home Team Name': 'Home_Team_Name', 'Away Team Name': 'Away_Team_Name'})
renamed_visualization1_df.columns = renamed_visualization1_df.columns.str.lower()
renamed_visualization1_df
```

```python
renamed_visualization1_df["stage_home_away_team"] = renamed_visualization1_df["home_team_name"] + " "+ renamed_visualization1_df["away_team_name"]+ " "+ renamed_visualization1_df["stage"]
renamed_visualization1_df.columns = renamed_visualization1_df.columns.str.lower()
renamed_visualization1_df

renamed_visualization1_df["unique"] = renamed_visualization1_df["stage_home_away_team"] + " "+ renamed_visualization1_df["year"].astype(str)
renamed_visualization1_df = renamed_visualization1_df.drop_duplicates(subset=['unique'])
renamed_visualization1_df = renamed_visualization1_df.drop(columns=["unique"])
renamed_visualization1_df
```

```python
# Group data by MatchID
grouped_visualization1_df = renamed_visualization1_df.groupby(["matchid"]).sum()
grouped_visualization1_df
```

```python
# Create dataframe that shows the Year, Home Team Goals, Away Team Goals and City
visualization2_df = soccer1_df[["Year", "Home Team Goals", "Away Team Goals", "City"]].dropna()
visualization2_df
```

```python
# visualization2_df.dtypes
visualization2_df['Total Goals'] = visualization2_df['Home Team Goals'] + visualization2_df['Away Team Goals'] 
visualization2_df
```

```python
# Rename columns
renamed_visualization2_df = visualization2_df.rename(columns={'Home Team Goals': 'home_team_goals', 'Away Team Goals': 'away_team_goals', 'Total Goals': 'Total_Goals'})
renamed_visualization2_df.columns = renamed_visualization2_df.columns.str.lower()
renamed_visualization2_df
```

```python
# Group data by year and city
grouped_visualization2_df = renamed_visualization2_df.groupby(["year", "city"]).sum().reset_index()
grouped_visualization2_df

# Create a new column in the DataFrame for latitude
grouped_visualization2_df['latitude'] = None

# Create a new column in the DataFrame for longitude
grouped_visualization2_df['longitude'] = None


grouped_visualization2_df
```

```python
cities = grouped_visualization2_df[["city","latitude","longitude"]]
cities = cities.drop_duplicates(subset=['city'])
cities
```

```python


# Create a new column in the DataFrame for latitude
grouped_visualization2_df['latitude'] = None

# Create a new column in the DataFrame for longitude
grouped_visualization2_df['longitude'] = None

# Create an instance of the geolocator
geolocator = GoogleV3(api_key=g_key)

# Iterate through the DataFrame and populate the latitude and longitude columns
for i, row in grouped_visualization2_df.iterrows():
    location = geolocator.geocode(row['city'])
    if location:
        grouped_visualization2_df.at[i, 'latitude'] = location.latitude
        grouped_visualization2_df.at[i, 'longitude'] = location.longitude
        
```
